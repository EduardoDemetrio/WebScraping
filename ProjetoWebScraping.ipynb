```python
### FUNAÇÃO DA TABELA BRASILEIRÃO MASCULINO
def tab_brasileirao_masc(tabelas_brasileiro_2023):
    Dados = []
    Ponto=[]
    Jogo = []
    Vitoria = []
    Empate = []
    Derrota=[]
    GolP =[]
    GolC =[]
    Sal=[]
    por=[]
    Ult_confron=[]
    CartAmarel=[]
    CartVerm=[]

    for i in range(1,41):
        Linha = tabelas_brasileiro_2023[0].find_all('tr')[i]

        ## Extraindo os pontos de todas as tabelas: 
        elementos_th = Linha.find_all('th')
        texto_th = [th.get_text() for th in elementos_th]
        Pontos = " ".join(texto_th)
        # print(Pontos)

        ## Extraindo os Times de todas as tabelas: 
        Time2 = Linha.find_all('td')[0].text.split('\n')
        Times2 = list(filter(lambda x: x.strip() != '', Time2))[-1]



        ## Extraindo os Pontos da tabela
        Jogos = ""
        if len(Linha.find_all('td')) >= 3:
            Jogos = Linha.find_all('td')[1].text

        ## Extraindo os Vitoria da tabela
        Vitorias = ""
        if len(Linha.find_all('td')) >= 3:
            Vitorias = Linha.find_all('td')[2].text

        ## Extraindo os empates da tabela
        Empates = ""
        if len(Linha.find_all('td')) >= 3:
            Empates = Linha.find_all('td')[3].text

        ## Extraindo os Derrotas da tabela 
        Derrotas = ""
        if len(Linha.find_all('td')) >= 3:
            Derrotas = Linha.find_all('td')[4].text


        ## Extraindo os Gols Pró da tabela 
        GP = ""
        if len(Linha.find_all('td')) >= 3:
            GP = Linha.find_all('td')[5].text

        ## Extraindo os Gols Contra da tabela 
        GC = ""
        if len(Linha.find_all('td')) >= 3:
            GC = Linha.find_all('td')[6].text

        ## Extraindo os Saldos Gol da tabela 
        SG = ""
        if len(Linha.find_all('td')) >= 3:
            SG = Linha.find_all('td')[7].text

        ## Extraindo os Cartão Amarelo da tabela 
        CA = ""
        if len(Linha.find_all('td')) >= 3:
            CA = Linha.find_all('td')[8].text

        ## Extraindo os Cartão Vermelho da tabela 
        CV = ""
        if len(Linha.find_all('td')) >= 3:
            CV = Linha.find_all('td')[9].text

        ## Extraindo os Aproveitamento da tabela 
        APROV = ""
        if len(Linha.find_all('td')) >= 3:
            APROV = Linha.find_all('td')[10].text

            # Golspartidas = Linha.find_all('td')[7].text
            # Golscontra= Linha.find_all('td')[8].text
            # Saldos= Linha.find_all('td')[9].text
            # porcent = Linha.find_all('td')[10].text

        Dados.append(Times2)
        Ponto.append(Pontos)
        Jogo.append(Jogos)
        Vitoria.append(Vitorias)
        Empate.append(Empates)
        Derrota.append(Derrotas)
        GolP.append(GP)
        GolC.append(GC)
        Sal.append(SG)
        por.append(APROV)
        CartAmarel.append(CA)
        CartVerm.append(CV)


    Times=[]
    for i in range(0, len(Dados), 2):
        time = Dados[i]
        Times.append(time)


    Ult_confron=[]
    for i in range(1, len(Dados), 2):
        time = Dados[i]
        Ult_confron.append(time)

    Jogos=[]
    for i in range(0, len(Jogo), 2):
        Jog = Jogo[i]
        Jogos.append(Jog)

    Pontos=[]
    for i in range(0, len(Ponto), 2):
        pont = Ponto[i]
        Pontos.append(pont)

    Vitorias=[]
    for i in range(0, len(Vitoria), 2):
        vit = Vitoria[i]
        Vitorias.append(vit)

    Empates=[]
    for i in range(0, len(Empate), 2):
        empt = Empate[i]
        Empates.append(empt)

    Derrotas=[]
    for i in range(0, len(Derrota), 2):
        derr = Derrota[i]
        Derrotas.append(derr)

    GolsP=[]
    for i in range(0, len(GolP), 2):
        gp = GolP[i]
        GolsP.append(gp)

    GolCont=[]
    for i in range(0, len(GolC), 2):
        gc = GolC[i]
        GolCont.append(gc)

    Saldo=[]
    for i in range(0, len(Sal), 2):
        sd = Sal[i]
        Saldo.append(sd)

    Aproveitamento=[]
    for i in range(0, len(por), 2):
        ap = por[i]
        Aproveitamento.append(ap)

    CartAmarelo =[]
    for i in range(0,len(CartAmarel),2):
        CA = CartAmarel[i]
        CartAmarelo.append(CA)

    CartVermelho =[]
    for i in range(0,len(CartVerm),2):
        Cv = CartVerm[i]
        CartVermelho.append(Cv)


    Dicionario = {
        'Time' : Times,
        'Pontos': Pontos,
        'Jogos': Jogos,
        'Vitorias':Vitorias,
        'Empates':Empates,
        'Derrotas':Derrotas,
        'Gols Pró' : GolsP,
        'Gols Contra': GolCont,
        'Saldos de gols':Saldo,
        'Porcentagem':Aproveitamento,
        'Cartões Amarelos': CartAmarelo,
        'Cartões Vermelhos':CartVermelho,
        'Ultimos resultados':Ult_confron}

    Tabela_brasileirao_2023 = pd.DataFrame(Dicionario)
    return Tabela_brasileirao_2023

### FUNÇÃO DA TABELA CATEGORIA DE BASE SUB 20 A E B
def tab_brasil_cat_base_a(tabelas_brasileiro_sub20):
    Dados = []
    Ponto=[]
    Jogo = []
    Vitoria = []
    Empate = []
    Derrota=[]
    GolP =[]
    GolC =[]
    Sal=[]
    por=[]
    Ult_confron=[]
    CartAmarel=[]
    CartVerm=[]

    for i in range(1,21):
        Linha = tabelas_brasileiro_sub20[0].find_all('tr')[i]

        ## Extraindo os pontos de todas as tabelas: 
        elementos_th = Linha.find_all('th')
        texto_th = [th.get_text() for th in elementos_th]
        Pontos = " ".join(texto_th)
        # print(Pontos)    

        # ## Extraindo os Times da tabelas sub 20: 
        Times3 = ""
        if len(Linha.find_all('td')) >= 3:
            Times3 = Linha.find_all('td')[0].text.split('\n')
            Times3 = list(filter(lambda x: x.strip() != '', Times3))[-1]

        # ## Extraindo os jogos da tabelas sub 20:
        Jogos2 = ""
        if len(Linha.find_all('td')) >= 3:
            Jogos2 = Linha.find_all('td')[1].text




        ## Extraindo os Vitoria da tabela do sub 20
        Vitorias = ""
        if len(Linha.find_all('td')) >= 3:
            Vitorias = Linha.find_all('td')[2].text

        ## Extraindo os empates da tabela
        Empates = ""
        if len(Linha.find_all('td')) >= 3:
            Empates = Linha.find_all('td')[3].text

        ## Extraindo os Derrotas da tabela 
        Derrotas = ""
        if len(Linha.find_all('td')) >= 3:
            Derrotas = Linha.find_all('td')[4].text


        ## Extraindo os Gols Pró da tabela 
        GP = ""
        if len(Linha.find_all('td')) >= 3:
            GP = Linha.find_all('td')[5].text

        ## Extraindo os Gols Contra da tabela 
        GC = ""
        if len(Linha.find_all('td')) >= 3:
            GC = Linha.find_all('td')[6].text

        ## Extraindo os Saldos Gol da tabela 
        SG = ""
        if len(Linha.find_all('td')) >= 3:
            SG = Linha.find_all('td')[7].text

        ## Extraindo os Cartão Amarelo da tabela 
        CA = ""
        if len(Linha.find_all('td')) >= 3:
            CA = Linha.find_all('td')[8].text

        ## Extraindo os Cartão Vermelho da tabela 
        CV = ""
        if len(Linha.find_all('td')) >= 3:
            CV = Linha.find_all('td')[9].text

        ## Extraindo os Aproveitamento da tabela 
        APROV = ""
        if len(Linha.find_all('td')) >= 3:
            APROV = Linha.find_all('td')[10].text

            # Golspartidas = Linha.find_all('td')[7].text
            # Golscontra= Linha.find_all('td')[8].text
            # Saldos= Linha.find_all('td')[9].text
            # porcent = Linha.find_all('td')[10].text

        Dados.append(Times3)
        Ponto.append(Pontos)
        Jogo.append(Jogos2)
        Vitoria.append(Vitorias)
        Empate.append(Empates)
        Derrota.append(Derrotas)
        GolP.append(GP)
        GolC.append(GC)
        Sal.append(SG)
        por.append(APROV)
        CartAmarel.append(CA)
        CartVerm.append(CV)

    Times=[]
    for i in range(0, len(Dados), 2):
        time = Dados[i]
        Times.append(time)


    Ult_confron=[]
    for i in range(1, len(Dados), 2):
        time = Dados[i]
        Ult_confron.append(time)

    Jogos=[]
    for i in range(0, len(Jogo), 2):
        Jog = Jogo[i]
        Jogos.append(Jog)

    Pontos=[]
    for i in range(0, len(Ponto), 2):
        pont = Ponto[i]
        Pontos.append(pont)

    Vitorias=[]
    for i in range(0, len(Vitoria), 2):
        vit = Vitoria[i]
        Vitorias.append(vit)

    Empates=[]
    for i in range(0, len(Empate), 2):
        empt = Empate[i]
        Empates.append(empt)

    Derrotas=[]
    for i in range(0, len(Derrota), 2):
        derr = Derrota[i]
        Derrotas.append(derr)

    GolsP=[]
    for i in range(0, len(GolP), 2):
        gp = GolP[i]
        GolsP.append(gp)

    GolCont=[]
    for i in range(0, len(GolC), 2):
        gc = GolC[i]
        GolCont.append(gc)

    Saldo=[]
    for i in range(0, len(Sal), 2):
        sd = Sal[i]
        Saldo.append(sd)

    Aproveitamento=[]
    for i in range(0, len(por), 2):
        ap = por[i]
        Aproveitamento.append(ap)

    CartAmarelo =[]
    for i in range(0,len(CartAmarel),2):
        CA = CartAmarel[i]
        CartAmarelo.append(CA)

    CartVermelho =[]
    for i in range(0,len(CartVerm),2):
        Cv = CartVerm[i]
        CartVermelho.append(Cv)


    Dicionario = {
        'Time' : Times,
        'Pontos': Pontos,
        'Jogos': Jogos,
        'Vitorias':Vitorias,
        'Empates':Empates,
        'Derrotas':Derrotas,
        'Gols Pró' : GolsP,
        'Gols Contra': GolCont,
        'Saldos de gols':Saldo,
        'Porcentagem':Aproveitamento,
        'Cartões Amarelos': CartAmarelo,
        'Cartões Vermelhos':CartVermelho,
        'Ultimos resultados':Ult_confron}

    Tabela_brasileirao_sub_20_2023 = pd.DataFrame(Dicionario)
    Tabela_brasileirao_sub_20_2023['Grupo de times'] = "Grupo A"
    return Tabela_brasileirao_sub_20_2023

def tab_brasil_cat_base_b(tabelas_brasileiro_sub20):
    Dados = []
    Ponto=[]
    Jogo = []
    Vitoria = []
    Empate = []
    Derrota=[]
    GolP =[]
    GolC =[]
    Sal=[]
    por=[]
    Ult_confron=[]
    CartAmarel=[]
    CartVerm=[]

    for i in range(1,21):
        Linha = tabelas_brasileiro_sub20[1].find_all('tr')[i]

        ## Extraindo os pontos de todas as tabelas: 
        elementos_th = Linha.find_all('th')
        texto_th = [th.get_text() for th in elementos_th]
        Pontos = " ".join(texto_th)
        # print(Pontos)    

        # ## Extraindo os Times da tabelas sub 20: 
        Times3 = ""
        if len(Linha.find_all('td')) >= 3:
            Times3 = Linha.find_all('td')[0].text.split('\n')
            Times3 = list(filter(lambda x: x.strip() != '', Times3))[-1]

        # ## Extraindo os jogos da tabelas sub 20:
        Jogos2 = ""
        if len(Linha.find_all('td')) >= 3:
            Jogos2 = Linha.find_all('td')[1].text




        ## Extraindo os Vitoria da tabela do sub 20
        Vitorias = ""
        if len(Linha.find_all('td')) >= 3:
            Vitorias = Linha.find_all('td')[2].text

        ## Extraindo os empates da tabela
        Empates = ""
        if len(Linha.find_all('td')) >= 3:
            Empates = Linha.find_all('td')[3].text

        ## Extraindo os Derrotas da tabela 
        Derrotas = ""
        if len(Linha.find_all('td')) >= 3:
            Derrotas = Linha.find_all('td')[4].text


        ## Extraindo os Gols Pró da tabela 
        GP = ""
        if len(Linha.find_all('td')) >= 3:
            GP = Linha.find_all('td')[5].text

        ## Extraindo os Gols Contra da tabela 
        GC = ""
        if len(Linha.find_all('td')) >= 3:
            GC = Linha.find_all('td')[6].text

        ## Extraindo os Saldos Gol da tabela 
        SG = ""
        if len(Linha.find_all('td')) >= 3:
            SG = Linha.find_all('td')[7].text

        ## Extraindo os Cartão Amarelo da tabela 
        CA = ""
        if len(Linha.find_all('td')) >= 3:
            CA = Linha.find_all('td')[8].text

        ## Extraindo os Cartão Vermelho da tabela 
        CV = ""
        if len(Linha.find_all('td')) >= 3:
            CV = Linha.find_all('td')[9].text

        ## Extraindo os Aproveitamento da tabela 
        APROV = ""
        if len(Linha.find_all('td')) >= 3:
            APROV = Linha.find_all('td')[10].text

            # Golspartidas = Linha.find_all('td')[7].text
            # Golscontra= Linha.find_all('td')[8].text
            # Saldos= Linha.find_all('td')[9].text
            # porcent = Linha.find_all('td')[10].text

        Dados.append(Times3)
        Ponto.append(Pontos)
        Jogo.append(Jogos2)
        Vitoria.append(Vitorias)
        Empate.append(Empates)
        Derrota.append(Derrotas)
        GolP.append(GP)
        GolC.append(GC)
        Sal.append(SG)
        por.append(APROV)
        CartAmarel.append(CA)
        CartVerm.append(CV)


    Times=[]
    for i in range(0, len(Dados), 2):
        time = Dados[i]
        Times.append(time)


    Ult_confron=[]
    for i in range(1, len(Dados), 2):
        time = Dados[i]
        Ult_confron.append(time)

    Jogos=[]
    for i in range(0, len(Jogo), 2):
        Jog = Jogo[i]
        Jogos.append(Jog)

    Pontos=[]
    for i in range(0, len(Ponto), 2):
        pont = Ponto[i]
        Pontos.append(pont)

    Vitorias=[]
    for i in range(0, len(Vitoria), 2):
        vit = Vitoria[i]
        Vitorias.append(vit)

    Empates=[]
    for i in range(0, len(Empate), 2):
        empt = Empate[i]
        Empates.append(empt)

    Derrotas=[]
    for i in range(0, len(Derrota), 2):
        derr = Derrota[i]
        Derrotas.append(derr)

    GolsP=[]
    for i in range(0, len(GolP), 2):
        gp = GolP[i]
        GolsP.append(gp)

    GolCont=[]
    for i in range(0, len(GolC), 2):
        gc = GolC[i]
        GolCont.append(gc)

    Saldo=[]
    for i in range(0, len(Sal), 2):
        sd = Sal[i]
        Saldo.append(sd)

    Aproveitamento=[]
    for i in range(0, len(por), 2):
        ap = por[i]
        Aproveitamento.append(ap)

    CartAmarelo =[]
    for i in range(0,len(CartAmarel),2):
        CA = CartAmarel[i]
        CartAmarelo.append(CA)

    CartVermelho =[]
    for i in range(0,len(CartVerm),2):
        Cv = CartVerm[i]
        CartVermelho.append(Cv)


    Dicionario = {
        'Time' : Times,
        'Pontos': Pontos,
        'Jogos': Jogos,
        'Vitorias':Vitorias,
        'Empates':Empates,
        'Derrotas':Derrotas,
        'Gols Pró' : GolsP,
        'Gols Contra': GolCont,
        'Saldos de gols':Saldo,
        'Porcentagem':Aproveitamento,
        'Cartões Amarelos': CartAmarelo,
        'Cartões Vermelhos':CartVermelho,
        'Ultimos resultados':Ult_confron}

    Tabela_brasileirao_sub_20_2023 = pd.DataFrame(Dicionario)
    Tabela_brasileirao_sub_20_2023['Grupo de times'] = "Grupo B"
    return Tabela_brasileirao_sub_20_2023


### FUNÇÃO DA TABELA BRASILEIRÃO FEMININO
def tab_brasilao_fem(tabelas_brasileiro_fem):
    Dados = []
    Ponto=[]
    Jogo = []
    Vitoria = []
    Empate = []
    Derrota=[]
    GolP =[]
    GolC =[]
    Sal=[]
    por=[]
    Ult_confron=[]
    CartAmarel=[]
    CartVerm=[]

    for i in range(1,33):
        Linha = tabelas_brasileiro_fem[0].find_all('tr')[i]

        ## Extraindo os pontos de todas as tabelas: 
        elementos_th = Linha.find_all('th')
        texto_th = [th.get_text() for th in elementos_th]
        Pontos = " ".join(texto_th)
        # print(Pontos)    

        # ## Extraindo os Times da tabelas sub 20: 
        Times3 = ""
        if len(Linha.find_all('td')) >= 3:
            Times3 = Linha.find_all('td')[0].text.split('\n')
            Times3 = list(filter(lambda x: x.strip() != '', Times3))[-1]

        # ## Extraindo os jogos da tabelas sub 20:
        Jogos2 = ""
        if len(Linha.find_all('td')) >= 3:
            Jogos2 = Linha.find_all('td')[1].text




        ## Extraindo os Vitoria da tabela do sub 20
        Vitorias = ""
        if len(Linha.find_all('td')) >= 3:
            Vitorias = Linha.find_all('td')[2].text

        ## Extraindo os empates da tabela
        Empates = ""
        if len(Linha.find_all('td')) >= 3:
            Empates = Linha.find_all('td')[3].text

        ## Extraindo os Derrotas da tabela 
        Derrotas = ""
        if len(Linha.find_all('td')) >= 3:
            Derrotas = Linha.find_all('td')[4].text


        ## Extraindo os Gols Pró da tabela 
        GP = ""
        if len(Linha.find_all('td')) >= 3:
            GP = Linha.find_all('td')[5].text

        ## Extraindo os Gols Contra da tabela 
        GC = ""
        if len(Linha.find_all('td')) >= 3:
            GC = Linha.find_all('td')[6].text

        ## Extraindo os Saldos Gol da tabela 
        SG = ""
        if len(Linha.find_all('td')) >= 3:
            SG = Linha.find_all('td')[7].text

        ## Extraindo os Cartão Amarelo da tabela 
        CA = ""
        if len(Linha.find_all('td')) >= 3:
            CA = Linha.find_all('td')[8].text

        ## Extraindo os Cartão Vermelho da tabela 
        CV = ""
        if len(Linha.find_all('td')) >= 3:
            CV = Linha.find_all('td')[9].text

        ## Extraindo os Aproveitamento da tabela 
        APROV = ""
        if len(Linha.find_all('td')) >= 3:
            APROV = Linha.find_all('td')[10].text

            # Golspartidas = Linha.find_all('td')[7].text
            # Golscontra= Linha.find_all('td')[8].text
            # Saldos= Linha.find_all('td')[9].text
            # porcent = Linha.find_all('td')[10].text

        Dados.append(Times3)
        Ponto.append(Pontos)
        Jogo.append(Jogos2)
        Vitoria.append(Vitorias)
        Empate.append(Empates)
        Derrota.append(Derrotas)
        GolP.append(GP)
        GolC.append(GC)
        Sal.append(SG)
        por.append(APROV)
        CartAmarel.append(CA)
        CartVerm.append(CV)


    Times=[]
    for i in range(0, len(Dados), 2):
        time = Dados[i]
        Times.append(time)


    Ult_confron=[]
    for i in range(1, len(Dados), 2):
        time = Dados[i]
        Ult_confron.append(time)

    Jogos=[]
    for i in range(0, len(Jogo), 2):
        Jog = Jogo[i]
        Jogos.append(Jog)

    Pontos=[]
    for i in range(0, len(Ponto), 2):
        pont = Ponto[i]
        Pontos.append(pont)

    Vitorias=[]
    for i in range(0, len(Vitoria), 2):
        vit = Vitoria[i]
        Vitorias.append(vit)

    Empates=[]
    for i in range(0, len(Empate), 2):
        empt = Empate[i]
        Empates.append(empt)

    Derrotas=[]
    for i in range(0, len(Derrota), 2):
        derr = Derrota[i]
        Derrotas.append(derr)

    GolsP=[]
    for i in range(0, len(GolP), 2):
        gp = GolP[i]
        GolsP.append(gp)

    GolCont=[]
    for i in range(0, len(GolC), 2):
        gc = GolC[i]
        GolCont.append(gc)

    Saldo=[]
    for i in range(0, len(Sal), 2):
        sd = Sal[i]
        Saldo.append(sd)

    Aproveitamento=[]
    for i in range(0, len(por), 2):
        ap = por[i]
        Aproveitamento.append(ap)

    CartAmarelo =[]
    for i in range(0,len(CartAmarel),2):
        CA = CartAmarel[i]
        CartAmarelo.append(CA)

    CartVermelho =[]
    for i in range(0,len(CartVerm),2):
        Cv = CartVerm[i]
        CartVermelho.append(Cv)


    Dicionario = {
        'Time' : Times,
        'Pontos': Pontos,
        'Jogos': Jogos,
        'Vitorias':Vitorias,
        'Empates':Empates,
        'Derrotas':Derrotas,
        'Gols Pró' : GolsP,
        'Gols Contra': GolCont,
        'Saldos de gols':Saldo,
        'Porcentagem':Aproveitamento,
        'Cartões Amarelos': CartAmarelo,
        'Cartões Vermelhos':CartVermelho,
        'Ultimos resultados':Ult_confron}

    Tabela_brasileirao_sub_20_2023 = pd.DataFrame(Dicionario)
    return Tabela_brasileirao_sub_20_2023


### FUNÇÃO DA TABELA DE ARTILHARIA DOS CAMPEONATOS
def artilheiros_as_camp(artilheiro_2023):
    QtdGol = []
    Nomes=[]
    Apelidos = []
    Times = []

    for i in range(1,16):
        Linha = artilheiro_2023[0].find_all('tr')[i]

        ## Extraindo quantidade de gols no ano: 
        elementos_th = Linha.find_all('th')
        texto_th = [th.get_text() for th in elementos_th]
        gol = " ".join(texto_th)
        # print(Pontos)

        ## Extraindo o Nome do jogador: 
        Name = Linha.find_all('td')[1].text


        ## Extraindo o Apelido do jogador:
        Apelido = ""
        if len(Linha.find_all('td')) >= 3:
            Apelido = Linha.find_all('td')[2].text

        Nomes.append(Name)
        QtdGol.append(gol)
        Apelidos.append(Apelido)


    for i in range(0,15):
        Linha = artilheiro_2023[0].find_all('img')[i]
        ## Extraindo o time do jogador: 
        Time = Linha['alt']
        Times.append(Time)



    Dicionario = {
        'Times' : Times,
        'Nome dos jogadores': Nomes,
        'Apelidos dos jogadores': Apelidos,
        'Quantos gols foram marcados no campeonato':QtdGol}

    Tabela_artilharia_masc = pd.DataFrame(Dicionario)
    return Tabela_artilharia_masc

```

## Teste de Duas variância

#### H0 : Variância de 2023 é igual a vairância de 2019
#### H1 : Variância de 2023 é menor que a vairância de 2019

```python
from scipy.stats import levene

# Duas amostras de dados
amostra1 = Tabela_brasileirao_2023['Média de gols']
amostra2 = Tabela_brasileirao_2019['Média de gols']

# Teste de Levene para igualdade de variâncias
stat, p_valor = levene(amostra1, amostra2)

print(f'Estatística do teste: {stat}')
print(f'Valor p: {p_valor}')

# Tomar decisão com base no valor p
nivel_significancia = 0.05
if p_valor > nivel_significancia:
    print('Não rejeitamos a hipótese nula - as variâncias são iguais.')
else:
    print('Rejeitamos a hipótese nula - há evidências de que as variâncias são diferentes.')
```
